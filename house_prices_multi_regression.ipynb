```python
import pandas as pd
```

```python
# Step 2 : import data
house = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/House%20Prices.csv')
```

```python
house.info()
```

```python
house.head()
```

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()


house['Date']= label.fit_transform(house['Date'])

#house['Date'] = house['Date'].astype(int, )
```

```python
house.head()
```

```python
house.describe()
```

```python
house.columns
```

```python
y = house['Price']
```

```python
X = house.drop(['Price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[0]
```

```python
X_test_single = X_test.iloc(0)

X_test_single = X_test.iloc[[0]]
X_test_single.shape

print(model.predict(X_test_single))

```

```python
# r2 value
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print(r2)
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
